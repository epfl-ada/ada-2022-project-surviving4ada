```python
#Sum the google trends indices
trend_2011.loc['Year_2011'] = trend_2011.apply(lambda x:x.sum())
trend_2012.loc['Year_2012'] = trend_2012.apply(lambda x:x.sum())
trend_2013.loc['Year_2013'] = trend_2013.apply(lambda x:x.sum())
trend_2011 = trend_2011[-1:]
trend_2012 = trend_2012[-1:]
trend_2013 = trend_2013[-1:]
```

```python
#Get the words that appear in all 3 years' top 100
list_all = []
for i in list_2011:
    if(i in list_2012)&(i in list_2013):
        list_all.append(i)
list_all = [s.replace('_',' ') for s in list_all]
print(len(list_all))
```

```python
#Apply some processes on the dataset
trend_2011_sorted = trend_2011.transpose().reset_index().sort_values(by='Year_2011', ascending = False).reset_index(drop = True)
trend_2012_sorted = trend_2012.transpose().reset_index().sort_values(by='Year_2012', ascending = False).reset_index(drop = True)
trend_2013_sorted = trend_2013.transpose().reset_index().sort_values(by='Year_2013', ascending = False).reset_index(drop = True)
#Select the 79 words that exist in all 3 years
trend_2011_sorted_new = trend_2011_sorted[trend_2011_sorted["index"].isin(list_all)].reset_index(drop=True)
trend_2012_sorted_new = trend_2012_sorted[trend_2012_sorted["index"].isin(list_all)].reset_index(drop=True)
trend_2013_sorted_new = trend_2013_sorted[trend_2013_sorted["index"].isin(list_all)].reset_index(drop=True)
trend_2011_sorted_new.rename(columns={'index':'Hub'}, inplace = True)
trend_2012_sorted_new.rename(columns={'index':'Hub'}, inplace = True)
trend_2013_sorted_new.rename(columns={'index':'Hub'}, inplace = True)
```

```python
#Add a new column "Google_trend_201X" to 3 years' dataset to represent their rankings
rank_list = []
for i in range(1,len(trend_2013_sorted_new)+1):
    rank_list.append(i)
trend_2011_sorted_new["Google_trend_2011"] = rank_list
trend_2012_sorted_new["Google_trend_2012"] = rank_list
trend_2013_sorted_new["Google_trend_2013"] = rank_list
trend_2011_sorted_new=trend_2011_sorted_new.drop(columns= "Year_2011")
trend_2012_sorted_new=trend_2012_sorted_new.drop(columns= "Year_2012")
trend_2013_sorted_new=trend_2013_sorted_new.drop(columns= "Year_2013")
```

```python
trend_1_2 = trend_2011_sorted_new.merge(trend_2012_sorted_new)
trend_3_years = trend_1_2.merge(trend_2013_sorted_new)
trend_3_years
```

```python
#Get the changes in rankings in 3 years
trend_3_years["Google_trend_change_1"] = trend_3_years["Google_trend_2011"] - trend_3_years["Google_trend_2012"]
trend_3_years["Google_trend_change_2"] = trend_3_years["Google_trend_2012"] - trend_3_years["Google_trend_2013"]
trend_3_years
```

```python
#Calculating the 79 words relative rankings in the wikispeedia datasets
TOP100_2011_wordlist = []
TOP100_2012_wordlist = []
TOP100_2013_wordlist = []
for i in range(0,100):
    TOP100_2011_wordlist.append(TOP100_2011[i][0])
    TOP100_2012_wordlist.append(TOP100_2012[i][0])
    TOP100_2013_wordlist.append(TOP100_2013[i][0])
TOP100_2011_wordlist = [s.replace("_", " ") for s in TOP100_2011_wordlist]
TOP100_2012_wordlist = [s.replace("_", " ") for s in TOP100_2012_wordlist]
TOP100_2013_wordlist = [s.replace("_", " ") for s in TOP100_2013_wordlist]
cnt = 0
order_2011 = {}
order_2012 = {}
order_2013 = {}
for i in TOP100_2011_wordlist:
    if i in list_all:
        cnt = cnt + 1
        order_2011[i] = cnt
cnt = 0
for i in TOP100_2012_wordlist:
    if i in list_all:
        cnt = cnt + 1
        order_2012[i] = cnt
cnt = 0
for i in TOP100_2013_wordlist:
    if i in list_all:
        cnt = cnt + 1
        order_2013[i] = cnt

#Get the rank
rank_list = []
for i in list_all:
    rank_list.append([i, order_2011[i],order_2012[i],order_2013[i]])
common_df = pd.DataFrame(rank_list, columns=['Hub','wiki_2011','wiki_2012','wiki_2013'])
common_df
```

```python
#Get the changes in rankings in 3 years
common_df["wiki_change_1"] = common_df["wiki_2011"] - common_df["wiki_2012"]
common_df["wiki_change_2"] = common_df["wiki_2012"] - common_df["wiki_2013"]
common_df
```

```python
#Merge the two dataframe
Union_df = trend_3_years.merge(common_df)
Union_df
```

```python
Union_df["Google_trend_change_1"] - Union_df['wiki_change_1']
```

```python
import numpy as np
#Calculate the error between the changes in different datasets 
diff_1 = np.sqrt(np.square(Union_df["Google_trend_2011"] - Union_df['wiki_2011']).sum()/2/79)
print("Rank mean square error in 2011:", diff_1)
diff_2 = np.sqrt(np.square(Union_df["Google_trend_2012"] - Union_df['wiki_2012']).sum()/2/79)
print("Rank mean square error in 2012:", diff_2)
diff_3 = np.sqrt(np.square(Union_df["Google_trend_2013"] - Union_df['wiki_2013']).sum()/2/79)
print("Rank mean square error in 2013:", diff_3)
error_1 = np.sqrt(np.square(Union_df["Google_trend_change_1"] - Union_df['wiki_change_1']).sum()/2/79)
print("Change mean square error_1:", error_1)
error_2 = np.sqrt(np.square(Union_df["Google_trend_change_2"] - Union_df['wiki_change_2']).sum()/2/79)
print("Change mean square error_2:",error_2)
```

```python
avg_delta_1 = np.mean(np.abs(Union_df["Google_trend_change_1"] - Union_df['wiki_change_1']))
avg_delta_2 = np.mean(np.abs(Union_df["Google_trend_change_2"] - Union_df['wiki_change_2']))
```

## Analysis:

<font size = 4>From the Union_df, we can see that the words' rankings in two datasets are sharply different. In order to prove this intuition, we calculated the 5 mean square errors. We can see that the mean square errors are very large. We can make conclusions that when people made choices on the wikispeedia navigations, the popularity of the word at that time did not make people tend to choose them in the navigation process. Therefore, the "tranport hubs" named by us are not easy to be influenced by the heated topics.
